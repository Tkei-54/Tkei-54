## The Artist's Canvas: Understanding Matplotlib's Structure

Matplotlib works like a digital art studio:

- **Figure**: Your canvas where everything is drawn
- **Axes**: Individual painting areas on your canvas
- **Plots**: The actual drawings (lines, points, bars) on your axes

This structure gives you incredible flexibility to create exactly the visualization you need.

# How Matplotlib Works: A Beginner's Guide

Matplotlib is a powerful library for creating a variety of plots in Python. It works in a layered mannerâ€”much like drawing on a canvas. Here's a simple breakdown of its main components:

## 1. Figure
- The **figure** is your blank canvas where all the plotting happens.
- You create a figure using `plt.figure()` or `plt.subplots()`.
- You can adjust the size of the figure with the `figsize` parameter.

## 2. Axes
- **Axes** are the areas within the figure where your data is plotted (they are essentially your subplots).
- A figure can have one or more axes.
- You can add axes using `fig.add_subplot()` or by creating them with `plt.subplots()`.

## 3. Plotting Data
- Once you have an axes, you can plot your data using functions like `ax.plot()`, `ax.scatter()`, or `ax.hist()`.
- This is where you visualize your data (e.g., line plots, scatter plots, histograms).

## 4. Customization
- **Titles and Labels:**  
  Use `ax.set_title()`, `ax.set_xlabel()`, and `ax.set_ylabel()` to add a title and axis labels.
- **Grid:**  
  You can add grid lines with `ax.grid()`, which help in reading the plot.
- **Legends, Colors, and Styles:**  
  Customize your plot further by setting colors, line styles, markers, and adding a legend if needed.

## 5. Displaying the Plot
- Finally, use `plt.show()` to render and display your figure.

## Line Plots: Tracking Trends Over Time

Line plots are perfect for showing how values change over a continuous variable (often time). Think of them as connecting the dots to show the journey of your data.


````python
# Generate sample data - 30 days of stock prices
days = np.arange(1, 31)
stock_price = 100 + np.cumsum(np.random.normal(0, 2, 30))  # Random walk starting at 100

# Output the data being used
print("Days:", days)
print("Stock Prices:", stock_price)

# Create a line plot
plt.figure(figsize=(12, 6))
plt.plot(days, stock_price, color='blue', linewidth=2, marker='o', markersize=6)

# Add title and labels
plt.title('Stock Price Over 30 Days', fontsize=16)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)

# Add a grid to make it easier to read values
plt.grid(True, linestyle='dashdot', alpha=0.7)

# Display the plot
plt.show()
````

np.random.normal(0, 2, 30):
- This generates 30 random numbers from a normal (Gaussian) distribution.
- The mean is 0, and the standard deviation is 2. This means that, on average, the values are centered around 0 with most values within a range of about Â±2.
- These numbers represent small, random changes (increments or decrements) in stock price.

np.cumsum(...):
- np.cumsum stands for cumulative sum.
- It takes the array of 30 random numbers and computes a running total. For example, if the random changes are [a, b, c, ...], the cumulative sum would be [a, a+b, a+b+c, ...].
- This cumulative process simulates a random walk where each step depends on the previous one.

100 + ...:
- By adding 100 to the cumulative sum, we set the starting point of the random walk.
This means that the simulated stock price begins at 100, and then the subsequent prices are determined by adding the random changes over time.

## Customizing Axes: Focusing Your Story

Just as a photographer frames a shot to highlight what's important, you can adjust your plot's axes to focus on the most relevant data.

### Setting Axis Limits

Use `plt.xlim()` and `plt.ylim()` to "zoom in" or "zoom out" on your data.

````python
# Generate 100 evenly spaced values between 0 and 9.9
x = np.arange(0, 10, 0.1)  # Goes from 0 to 9.9, stepping exactly by 0.1

# Calculate y = xÂ² for each value in x
y = x ** 2

print(len(x))

print("X Values:", x)
print("Y Values:", y)

# Create a figure with 2 subplots arranged side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# ---------------------------
# First subplot: Default view
# ---------------------------
# Plot x vs. y with a solid red line ('r-')
ax1.plot(x, y, 'r-')
# Set the title for the first plot
ax1.set_title('Default View')
# Label the x-axis
ax1.set_xlabel('x')
# Label the y-axis showing the function y = xÂ²
ax1.set_ylabel('y = xÂ²')
# Add a grid to help read the plot values
ax1.grid(True)

# ---------------------------
# Second subplot: Zoomed in view
# ---------------------------
# Plot x vs. y with a solid red line ('r-')
ax2.plot(x, y, 'r-')
# Set the title for the second plot
ax2.set_title('Zoomed In View')
# Label the x-axis
ax2.set_xlabel('x')
# Label the y-axis showing the function y = xÂ²
ax2.set_ylabel('y = xÂ²')
# Limit the x-axis to values between 4 and 6 (zoom in)
ax2.set_xlim(4, 6)
# Limit the y-axis to values between 15 and 40 (zoom in)
ax2.set_ylim(15, 40)
# Add a grid to help read the plot values
ax2.grid(True)

# Adjust spacing between subplots to prevent overlap
plt.tight_layout()
# Display the plots
plt.show()
````

### Customizing Axis Ticks

You can control where the tick marks appear on your axes and how they're labeled.

````python
# Generate data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y)

# Set custom x-ticks at multiples of Ï€
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           ['0', '90', '180', '270', '360'])

# Set custom y-ticks
plt.yticks([-1, -0.5, 0, 0.5, 1],
           ['Minimum', '-0.5', '0.0', '0.5', 'Maximum'])

plt.title('Sine Wave with Custom Tick Labels')
plt.grid(True)
plt.show()
````

## Customizing Line Styles: Adding Visual Distinction

When plotting multiple lines, different styles help distinguish between them. Matplotlib offers a variety of options for customizing your lines.

````python
# Generate data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

# Create the plot
plt.figure(figsize=(12, 6))

# Plot with different line styles
plt.plot(x, y1, 'b-', linewidth=2, label='Solid line (sin(x))')
plt.plot(x, y2, 'r--', linewidth=2, label='Dashed line (cos(x))')
#plt.plot(x, y2, color='red', linestyle='--', linewidth=2, label='Dashed line (cos(x))')
plt.plot(x, y3, 'g-.', linewidth=2, label='Dash-dot line (sin(x)cos(x))')

# Add title and labels
plt.title('Different Line Styles', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)

# Add a grid and legend
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12, loc = 'upper center')

# Display the plot
plt.show()
````

### Line Style Reference

Here's a quick reference for common line styles in Matplotlib:

- Solid line: `-` or `solid`
- Dashed line: `--` or `dashed`
- Dash-dot line: `-.` or `dashdot`
- Dotted line: `:` or `dotted`

You can also customize:
- Line width with `linewidth` or `lw`
- Line color with color names (`'red'`), hex codes (`'#FF0000'`), or RGB tuples (`(1.0, 0.0, 0.0)`)
- Markers with `marker` (e.g., `'o'` for circles, `'s'` for squares, `'^'` for triangles)

## Multiple Plots on the Same Axes

Often, you'll want to compare multiple data series on the same plot. This is easy to do in Matplotlib.

````python
# Generate data for three cities' temperatures
months = np.arange(1, 13)
city1_temps = [5, 7, 10, 15, 20, 25, 28, 27, 22, 15, 10, 6]  # Northern city
city2_temps = [20, 22, 25, 27, 29, 30, 31, 30, 28, 25, 22, 20]  # Tropical city
city3_temps = [25, 24, 22, 18, 15, 12, 10, 12, 15, 18, 20, 23]  # Southern hemisphere city

# Create the plot
plt.figure(figsize=(12, 6))

# Plot all three temperature series
plt.plot(months, city1_temps, 'b-o', linewidth=2, label='Northern City')
plt.plot(months, city2_temps, 'r-^', linewidth=2, label='Tropical City')
plt.plot(months, city3_temps, 'g-s', linewidth=2, label='Southern City')

# Add title and labels
plt.title('Average Monthly Temperatures', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Temperature (Â°C)', fontsize=12)

# Set the x-ticks to month names
plt.xticks(months, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Add a grid and legend
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

# Display the plot
plt.show()
````

## Creating Subplots: Multiple Views in One Figure

Subplots allow you to create multiple plots within a single figure. This is useful for comparing different aspects of your data or showing related visualizations together.

### Real-world Analogy
Think of subplots like a comic strip or a photo collage - each panel tells part of a larger story.

````python
# Generate data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)
y4 = np.sin(x) + np.cos(x)

# Create a figure with 2 rows and 2 columns of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot on each subplot
axs[0, 0].plot(x, y1, 'b-')
axs[0, 0].set_title('Sine Wave')
axs[0, 0].set_ylabel('sin(x)')
axs[0, 0].grid(True)

axs[0, 1].plot(x, y2, 'r-')
axs[0, 1].set_title('Cosine Wave')
axs[0, 1].set_ylabel('cos(x)')
axs[0, 1].grid(True)

axs[1, 0].plot(x, y3, 'g-')
axs[1, 0].set_title('Sine Ã— Cosine')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('sin(x) Ã— cos(x)')
axs[1, 0].grid(True)

axs[1, 1].plot(x, y4, 'm-')
axs[1, 1].set_title('Sine + Cosine')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('sin(x) + cos(x)')
axs[1, 1].grid(True)

# Add a title to the entire figure
fig.suptitle('Trigonometric Functions', fontsize=16)

# Adjust spacing between subplots
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Make room for the suptitle

# Display the plot
plt.show()
````

### Creating Subplots with Different Sizes

Sometimes you want some subplots to be larger than others. You can achieve this using `gridspec`.

````python
import matplotlib.gridspec as gridspec

# Create a figure
fig = plt.figure(figsize=(12, 8))

# Create a grid with 2 rows and 3 columns
gs = gridspec.GridSpec(2, 3)

# Create subplots with different sizes
ax1 = fig.add_subplot(gs[0, :])  # Top row, spans all columns
ax2 = fig.add_subplot(gs[1, 0])  # Bottom row, first column
ax3 = fig.add_subplot(gs[1, 1:])  # Bottom row, spans second and third columns

# Add data to each subplot
ax1.plot(np.random.rand(100), 'b-')
ax1.set_title('Main Time Series')
ax1.set_ylabel('Value')
ax1.grid(True)

ax2.hist(np.random.normal(0, 1, 1000), bins=20, color='green', alpha=0.7)
ax2.set_title('Distribution')
ax2.set_xlabel('Value')
ax2.set_ylabel('Frequency')
ax2.grid(True)

ax3.scatter(np.random.rand(50), np.random.rand(50), color='red', alpha=0.7)
ax3.set_title('Scatter Plot')
ax3.set_xlabel('X Value')
ax3.set_ylabel('Y Value')
ax3.grid(True)

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
````

## Object-Oriented vs. PyPlot Syntax

Matplotlib offers two main approaches to creating visualizations: the object-oriented interface and the PyPlot interface. Let's compare them:

````python
# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# PyPlot interface (state-machine style)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b-')
plt.title('PyPlot Interface')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)

# Object-oriented interface
fig, ax = plt.subplots(1, 1, figsize=(6, 5))
ax.plot(x, y, 'r-')
ax.set_title('Object-Oriented Interface')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.grid(True)

plt.tight_layout()
plt.show()
````

Think of PyPlot as a whiteboard.

- You draw directly on the whiteboard, and each new command affects the entire board.
- It automatically keeps track of the current figure and axes.
- You donâ€™t need to manually manage objects like figures or axes.

Now think of Object-Oriented Matplotlib as a canvas on an easel.

- You create the canvas (Figure) and then draw on specific parts (Axes).
- Instead of affecting the entire figure at once, you control each subplot independently.
- More powerful and better for complex plots.

### Which Approach Should You Use?

**PyPlot Interface:**
- Pros: Simpler for quick, one-off plots; fewer lines of code
- Cons: Less flexible for complex plots; harder to customize

**Object-Oriented Interface:**
- Pros: More flexible; better for complex plots; clearer code for multiple subplots
- Cons: Slightly more verbose for simple plots

**Recommendation:** For beginners, the PyPlot interface is fine for simple plots. As you create more complex visualizations, the object-oriented approach becomes more valuable. Most professional data scientists prefer the object-oriented approach for its flexibility and clarity.

## Advanced Customization: Annotations and Text

Adding text annotations to your plots can help highlight important points or provide additional context.

````python
# Generate data for a stock price
days = np.arange(1, 31)
np.random.seed(42)  # For reproducibility
stock_price = 100 + np.cumsum(np.random.normal(0, 2, 30))  # Random walk starting at 100

# Find the highest and lowest points
max_idx = np.argmax(stock_price)
min_idx = np.argmin(stock_price)

# Create the plot
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(days, stock_price, 'b-', linewidth=2)

# Add annotations for maximum and minimum points
ax.annotate(f'Peak: ${stock_price[max_idx]:.2f}',
            xy=(days[max_idx], stock_price[max_idx]),
            xytext=(days[max_idx]+2, stock_price[max_idx]+5),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5),
            fontsize=12)

ax.annotate(f'Low: ${stock_price[min_idx]:.2f}',
            xy=(days[min_idx], stock_price[min_idx]),
            xytext=(days[min_idx]+2, stock_price[min_idx]-5),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5),
            fontsize=12)

# Add a text box with analysis
textstr = '\n'.join((
    'Stock Analysis:',
    f'Starting Price: ${stock_price[0]:.2f}',
    f'Ending Price: ${stock_price[-1]:.2f}',
    f'Change: {(stock_price[-1] - stock_price[0]):.2f} ({(stock_price[-1] - stock_price[0])/stock_price[0]*100:.1f}%)'))

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=12,
        verticalalignment='top', bbox=props)

# Add title and labels
ax.set_title('Stock Price Over 30 Days with Annotations', fontsize=16)
ax.set_xlabel('Day', fontsize=12)
ax.set_ylabel('Price ($)', fontsize=12)
ax.grid(True)

plt.tight_layout()
plt.show()
````

ax.annotate(text, xy, xytext, arrowprops, fontsize):
- Marks important points with annotations.
Parameters Explained:
- text â†’ Displays stock price (f'Peak: ${stock_price[max_idx]:.2f}').
- xy=(x, y) â†’ Location of the annotation (peak or lowest stock price).
- xytext=(x_offset, y_offset) â†’ Moves the text slightly away to improve readability.
- arrowprops=dict(...) â†’ Draws an arrow pointing to the exact price.
- facecolor='black' â†’ Arrow color.
- shrink=0.05 â†’ Shortens the arrow slightly.
- width=1.5 â†’ Makes the arrow thicker.
- fontsize=12 â†’ Adjusts text size.

textstr:
- Creates a summary text using f-strings (formatted strings).
Displays:
- Starting price
- Ending price
- Percentage change in stock value.
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5):
- Defines a styled text box.
- Rounded shape ('round').
- Wheat-colored background ('facecolor='wheat').
- Transparency (alpha=0.5) for better readability.
ax.text(...):
- Places the text box in the upper-left corner ((0.05, 0.95)) of the plot.
- Uses axis coordinates (transform=ax.transAxes), so it always stays in the corner.
